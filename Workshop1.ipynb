```python
#problem 4

import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

# 1. Prove A * A‚Åª¬π = I
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("\n1. A * A‚Åª¬π = I:\n", I)
```

```python
# 2. Prove AB ‚â† BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("\nBA:\n", BA)
print("\nSince AB ‚â† BA ‚Üí Matrix multiplication is NOT commutative.")

```

```python
# 3. Prove (AB)·µÄ = B·µÄA·µÄ
AB = np.dot(A, B)
BA = np.dot(B, A)
LHS = (AB).T       # Transpose of AB
RHS = np.dot(B.T, A.T)  # B·µÄA·µÄ
print("\n3. (AB)·µÄ:\n", LHS)
print("\nB·µÄA·µÄ:\n", RHS)
print("\nTherefore, (AB)·µÄ = B·µÄA·µÄ ‚úî")
```

```python
# SOlve using inverse
# Coefficient matrix A and constants matrix B
A2 = np.array([[2, -3, 1],
               [1, -1, 2],
               [3, 1, -1]])

B2 = np.array([-1, -3, 9])

# Using inverse method ‚Üí X = A‚Åª¬πB
A2_inv = np.linalg.inv(A2)
solution = np.dot(A2_inv, B2)

print("\nSolution using Inverse Method:")
print("x, y, z =", solution)
```

```python
#Solve using np.linalg.solve() (Direct Method)

solution2 = np.linalg.solve(A2, B2)
print("\nSolution using numpy solve():")
print("x, y, z =", solution2)
```
